```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(text):
    L = []
    # Handle both raw strings and already-parsed lists
    items = text if isinstance(text, list) else ast.literal_eval(text)
    for i in items:
        # Handle both dicts {'name': '...'} and plain strings
        L.append(i['name'] if isinstance(i, dict) else i)
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    # Handle both string and already-parsed list
    items = text if isinstance(text, list) else ast.literal_eval(text)
    for i in items:
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```
